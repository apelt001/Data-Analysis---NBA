### Description of Columns

We have merged the data with data from the [nba_api](https://github.com/swar/nba_api), a very useful python wrapper for stats.nba.com that has a decent community slack channel. Here is a description of some of the merged columns.

- `Market_Full_Label` - Points, Rebounds, or Assists
- `Projection` - the site's own projected outcome.
- `Recommendation` - the site's recommendation (*over*/*under*) based on the difference between `Projection` and `Line`
- `rec_DIFF` - the difference between `Projection` and `Line`
- `Actual` - the actual outcome.
- `outcome_DIFF` - the difference between `Actual` and `Line`
- `hundred_rec` - profit (positive or negative) from a hundred dollar that follows the `Recommendation` column.

  The rest of the columns are stats. Stat names that are appended with `_PLAYER`, `_TEAM`, or `_OPP`. IE in the first row, John Collins had $2$ blocks, his team had $10$ total blocks, and the opposing team had $6$ total blocks.

```python
df.loc[:,['SEASON_YEAR','PLAYER_NAME','BLK_PLAYER','BLK_TEAM','BLK_OPP']].head()
```

What we want to compare is how well the `Projection` column performs on the data.

```python
df.groupby(['Recommendation','Market_Full_Label'])['hundred_rec'].agg(['count','mean'])
```

From this result, we see that the recommendations seem to be good for the **Points** and **Rebounds** whenever the recommendation is *under*, but NOT if the recommendation is *over* (The average ROI is $5.8$% on a Rebounds prop bet if the recommendation is *under*, and $-2.8$% if the recommendation is *over*).

```python
df.groupby('Recommendation')['hundred_rec'].agg(['count','mean'])
```

Overall, it seems that the **Recommendation** is only good when it is under.

```python
df.groupby(['Recommendation','Market_Full_Label','SEASON_YEAR'])['hundred_rec'].agg(['count','mean'])
```

Now we want to see if we should be *more confident* in a *projection* with a larger gap than a smaller one. To do this, we can use **pd.cut**.

```python
df['diff_bins'] = pd.cut(df['rec_DIFF'], bins = [-10,-3.8,-3.1,-2.5,-0.8,0,1.1,3.1,10])
```

```python
POINTS = df[df['Market_Full_Label'] == 'Points']
POINTS.groupby('diff_bins')['hundred_rec'].agg(['count','mean'])
```

As you can see with the **Points** prop bets, an **UNDER** recommendation is better when there is a large difference between the **Projection** column and the actual **Line**. However, an **OVER** recommendation is never good. The average ROI on a $100\$$ wager when there is a $-3.8$ point difference or less is $6.70\$$.

```python
df['diff_bins'] = pd.cut(df['rec_DIFF'], bins = [-10,-2.0,-1.1,0,1.1,3.1,10])
REBOUNDS = df[df['Market_Full_Label'] == 'Rebounds']
REBOUNDS.groupby('diff_bins')['hundred_rec'].agg(['count','mean'])
```

For Rebounds, the same trend is present; the strategy says *bet the UNDER* if the Betting Pros projection says so, and bet multiple units if the projection differs from the Line by $2$ rebounds or more. The strategy also says *never bet the OVER*.
